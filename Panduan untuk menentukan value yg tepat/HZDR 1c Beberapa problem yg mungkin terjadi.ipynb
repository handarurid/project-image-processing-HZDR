### Problem terkait dengan threshold yang kurang sesuai (garis permukaan air turun sampai ke dasar)
Threshold yang digunakan pada contoh ini adalah 210 (warna pixel di bawah 210 akan jadi hitam, di atas 210 akan jadi putih). Kalau ternyata garis permukaan air tidak terdeteksi, maka value 210 ini boleh dinaikkan sedikit demi sedikit agar warna gelap pada pixel permukaan air menjadi lebih banyak. Misal dinaikkan jadi 215, lalu kalau masih tidak terdeteksi bisa dinaikkan lagi jadi 220, dst.

Contoh kasusnya seperti ini:

```python
def preprocess_image(filename):
    bgr_image       = cv2.imread(filename)
    rgb_image       = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)
    gray_image      = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2GRAY)
    # ANGKA THRESHOLD ADALAH 210 YANG ADA DI BAWAH SINI
    _, binary_image = cv2.threshold(gray_image, 210, 255, cv2.THRESH_BINARY)
    binary_image    = binary_image/255
    
    return binary_image, gray_image
```
