```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# the input was a filename (now it's the image itself)
def preprocess_image(image):
    #image = cv2.imread(filename)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    _, image = cv2.threshold(image, 210, 255, cv2.THRESH_BINARY)
    binary_image = image / 255
    return binary_image
```

```python
cap = cv2.VideoCapture('CCFL002_1_2000_500_up_cam2_a.avi')

while(cap.isOpened()):
    ret, frame = cap.read()
    
    binary = preprocess_image(frame)

    cv2.imshow('frame',binary)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import pylab
import imageio
filename = 'CCFL002_1_2000_500_up_cam2_a.avi'
vid = imageio.get_reader(filename,  'ffmpeg')
nums = [10, 287]
for num in nums:
    image = vid.get_data(num)
    fig = pylab.figure()
    fig.suptitle('image #{}'.format(num), fontsize=20)
    pylab.imshow(image)
pylab.show()
```
