```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from scipy import signal
from scipy.stats import beta
```

```python
df = pd.read_csv('Data.csv')
df = df.iloc[:,1:]
```

```python
df.head()
```

```python
df['Point 1'].hist(bins=40)
plt.title('Histogram')
plt.show()
```

```python
df['Point 1'].plot.density()
plt.title('Density')
plt.show()
```

```python
np.log(df['Point 1']).plot.density()
plt.title('Log Density')
plt.show()
```

```python
plt.psd(df['Point 1'])
plt.title('PSD')
plt.show()
```

```python
cwtmatr = signal.cwt(df['Point 1'], signal.ricker, np.arange(1, 31))
```

```python
plt.imshow(cwtmatr, extent=[-1, 1, 1, 31], cmap='PRGn', aspect='auto',
           vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
plt.show()
```

## Yang bawah ini cuma buat testing pakai data dummy

```python
from scipy import signal
import matplotlib.pyplot as plt
t = np.linspace(-1, 1, 200, endpoint=False)
sig  = np.cos(2 * np.pi * 7 * t) + signal.gausspulse(t - 0.4, fc=2)
widths = np.arange(1, 31)
cwtmatr = signal.cwt(sig, signal.ricker, widths)
plt.imshow(cwtmatr, extent=[-1, 1, 1, 31], cmap='PRGn', aspect='auto',
           vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
plt.show()
```
