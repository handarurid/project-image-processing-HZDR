```python
import gc
import cv2
import imageio
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
# the input was a filename (now it's the image itself)
def preprocess_image(image):
    #image = cv2.imread(filename)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    _, image = cv2.threshold(image, 210, 255, cv2.THRESH_BINARY)
    binary_image = image / 255
    return binary_image
```

```python
filename = 'CCFL002_1_2000_500_up_cam2_a.avi'
video = imageio.get_reader(filename, 'ffmpeg')
```

```python
image = video.get_data(0)
```

```python
plt.imshow(image)
```

```python
frames = []
for i in tqdm(range(600)):
    image = np.array(video.get_data(i))
    image = preprocess_image(image)
    frames.append(np.array(image))

frames = np.array(frames)
```

```python
plt.imshow(frames[0])
```

```python
# Memory error!
frames = []
for i, frame in tqdm(enumerate(video)):
    image = np.array(video.get_data(i))
    image = preprocess_image(image)
    frames.append(np.array(image))
    gc.collect()
```
